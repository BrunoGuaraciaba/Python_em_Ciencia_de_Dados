````python
# 1) Seu c√≥digo aqui
#Cria uma c√≥pia da base de pagamentos
pagamentos_default = pg.copy()

#Colunas de meses (mes_00 at√© mes_24)
colunas_meses = [col for col in pagamentos_default.columns if 'mes_' in col] 

# Marca 1 para meses com atraso >= 60 dias
pagamentos_default[colunas_meses] = pagamentos_default[colunas_meses].isin(['2','3','4','5']).astype(int)

# Visualiza as primeiras linhas
pagamentos_default.head()

#Explica√ß√£o do codigo
# # Cria uma c√≥pia da base de pagamentos
# pagamentos_default = pagamentos.copy()
# Faz uma c√≥pia da base original. Isso evita modificar diretamente o DataFrame pagamentos ‚Äî uma boa pr√°tica para manter a base original intacta.

# # Colunas de meses (mes_00 at√© mes_24)
# colunas_meses = [col for col in pagamentos_default.columns if 'mes_' in col]

# Cria uma lista com os nomes das colunas que cont√™m ‚Äúmes_‚Äù, ou seja, todas as colunas de mes_00 a mes_24.
# Esse comando √© uma list comprehension (forma compacta de gerar listas).
#  Exemplo de resultado: ['mes_00', 'mes_01', ..., 'mes_24']
# Isso √© um loop (la√ßo de repeti√ß√£o) dentro de uma list comprehension.
# Ele vai percorrer cada coluna da lista pagamentos_default.columns, uma por uma.
# Em cada volta do la√ßo, a vari√°vel col assume o nome da coluna atual.
# Exemplo pr√°tico:
# Itera√ß√£o	Valor de col
# 1	'ID'
# 2	'mes_00'
# 3	'mes_01'
# ...	'mes_24'
    
# # Marca 1 para meses com atraso >= 60 dias
# pagamentos_default[colunas_meses] = pagamentos_default[colunas_meses].isin(['2','3','4','5']).astype(int)
# Explicando passo a passo:

# pagamentos_default[colunas_meses] ‚Üí seleciona apenas as colunas de meses.

# .isin(['2','3','4','5']) ‚Üí cria uma matriz de True/False, onde True indica que o valor da c√©lula est√° entre os c√≥digos '2', '3', '4', '5' (ou seja, atraso ‚â• 60 dias).

# .astype(int) ‚Üí converte True em 1 e False em 0.

# O resultado √© que, em cada coluna de m√™s, 1 significa ‚Äúcliente em default naquele m√™s‚Äù.

# # Visualiza as primeiras linhas
# pagamentos_default.head()
# Mostra as primeiras 5 linhas da base para confer√™ncia visual (fun√ß√£o de checagem).

````

````python
# 2) Seu c√≥digo aqui
# Soma dos meses com default
pagamentos_default['qtd_meses_default'] = pagamentos_default[colunas_meses].sum(axis=1)

# Vari√°vel final: 1 = Mau pagador, 0 = Bom pagador
pagamentos_default['default'] = (pagamentos_default['qtd_meses_default'] > 0).astype(int)

# Mostrando o resultado
pagamentos_default[['ID','default']].head()
````

````python
# 3) Seu c√≥digo aqui
Como o enunciado diz que a base de pagamentos cont√©m apenas clientes que receberam o cart√£o, n√£o √© preciso fazer uma filtragem adicional.
Na pr√≥xima etapa, ao fazer a jun√ß√£o (merge), s√≥ permanecer√£o os IDs que est√£o em ambas as tabelas.
````

````python
# 4) eu c√≥digo aqui
# Jun√ß√£o usando a chave ID
base_final = propostas.merge(
    pagamentos_default[['ID', 'default']],
    on='ID',
    how='inner'   # apenas quem possui informa√ß√µes de pagamento
)

base_final.head()
````
