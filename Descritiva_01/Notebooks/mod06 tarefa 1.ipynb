```python
import pandas as pd
import matplotlib.pyplot as plt

#Carregar base CSV
sinasc = pd.read_csv('Profissão Cientista de Dados M6 - SINASC_RO_2019.csv', sep=',')

#Visualizar primeiras linhas
print(sinasc.head())

#Verificar estrutura
print(sinasc.info())
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
# APGAR1
sinasc['APGAR1'].value_counts().sort_index().plot(kind='bar')
plt.title('Frequência APGAR1')
plt.xlabel('APGAR1')
plt.ylabel('Frequência')
plt.show()

# APGAR5
sinasc['APGAR5'].value_counts().sort_index().plot(kind='bar')
plt.title('Frequência APGAR5')
plt.xlabel('APGAR5')
plt.ylabel('Frequência')
plt.show()

# SEXO
sinasc['SEXO'].value_counts().plot(kind='bar')
plt.title('Frequência SEXO')
plt.xlabel('SEXO')
plt.ylabel('Frequência')
plt.show()

# GESTACAO
sinasc['GESTACAO'].value_counts().plot(kind='bar')
plt.title('Frequência GESTAÇÃO')
plt.xlabel('GESTAÇÃO')
plt.ylabel('Frequência')
plt.xticks(rotation=45)
plt.show()

# ESTCIVMAE
sinasc['ESTCIVMAE'].value_counts().plot(kind='bar')
plt.title('Frequência ESTADO CIVIL MÃE')
plt.xlabel('Estado Civil')
plt.ylabel('Frequência')
plt.xticks(rotation=45)
plt.show()

#sinasc['APGAR1']
#sinasc → é o DataFrame carregado do CSV
#['APGAR1'] → seleciona apenas a coluna APGAR1

#.value_counts() => Conta quantas vezes cada valor aparece.
#Ele cria uma nova Series onde:
#índice = valores de APGAR1
#valores = quantidade de ocorrências

#.sort_index() Ordena os valores pelo índice. Sem isso, o pandas ordena pela frequência (do maior para o menor).Com sort_index(), 
#ele ordena pelo valor do APGAR:
# 0, 1, 2, 3 ... 10


#.plot(kind='bar') => Cria o gráfico.
#kind='bar' → gráfico de barras
#O eixo X será o índice (valores do APGAR1)
#O eixo Y será a frequência
#
#
```

3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
apgar_consultas = sinasc.groupby('CONSULTAS')['APGAR5'].mean()

apgar_consultas.plot(kind='bar')
plt.title('Média APGAR5 por Número de Consultas')
plt.xlabel('Consultas')
plt.ylabel('Média APGAR5')
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# SEXO
sinasc['SEXO'].value_counts().plot(kind='bar', ax=axs[0,0])
axs[0,0].set_title('Frequência SEXO')

# GESTACAO
sinasc['GESTACAO'].value_counts().plot(kind='bar', ax=axs[0,1])
axs[0,1].set_title('Frequência GESTAÇÃO')

# ESTCIVMAE
sinasc['ESTCIVMAE'].value_counts().plot(kind='bar', ax=axs[1,0])
axs[1,0].set_title('Frequência ESTADO CIVIL MÃE')

# CONSULTAS
sinasc['CONSULTAS'].value_counts().plot(kind='bar', ax=axs[1,1])
axs[1,1].set_title('Frequência CONSULTAS')

plt.tight_layout()
plt.show()
```
