####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#Carregar base CSV
sinasc = pd.read_csv('Profissão Cientista de Dados M6 - SINASC_RO_2019.csv', sep=',')

#Visualizar primeiras linhas
print(sinasc.head())

#Verificar estrutura
print(sinasc.info())
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x=sinasc['IDADEMAE'].dropna()
)

plt.title('Boxplot - Idade da Mãe')
plt.xlabel('Idade da Mãe')

plt.show()


# Cria uma nova figura (janela do gráfico). figsize=(8,5) define largura e altura em polegadas.

# sns.boxplot() → função que cria boxplot

# x= → estamos fazendo boxplot horizontal

# sinasc['IDADEMAE'] → seleciona a coluna

# .dropna() → remove valores nulos

# Se não usar dropna(), pode gerar erro ou distorção.
```

####  2) Faça um boxplot do peso do bebê

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x=sinasc['PESO'].dropna()
)

plt.title('Boxplot - Peso do Bebê')
plt.xlabel('Peso (gramas)')

plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
plt.figure(figsize=(8,5))

sns.histplot(
    data=sinasc,
    x='IDADEMAE',
    bins=20,
    kde=True
)

plt.title('Histograma - Idade da Mãe')
plt.xlabel('Idade')
plt.ylabel('Frequência')

plt.show()

# data=sinasc

# Passa o DataFrame inteiro.

# x='IDADEMAE'

# Seleciona a coluna dentro do DataFrame.

# bins=20

# Divide os dados em 20 intervalos.
# Mais bins → mais detalhado
# Menos bins → mais agrupado

# kde=True

# Adiciona uma curva suavizada (estimativa de densidade).
# Mostra tendência da distribuição.
```

#### 4) Faça um histograma do peso do bebê

```python
plt.figure(figsize=(8,5))

sns.histplot(
    data=sinasc,
    x='PESO',
    bins=30,
    kde=True
)

plt.title('Histograma - Peso do Bebê')
plt.xlabel('Peso (gramas)')
plt.ylabel('Frequência')

plt.show()
```
